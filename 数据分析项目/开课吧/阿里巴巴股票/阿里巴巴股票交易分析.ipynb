```python
# loadtxt方法有一个converters参数, 可以利用自定义的函数把string做转换
def convert_date(d):
    return  parse(d).weekday()
stock_info = np.loadtxt(r'E:\项目\BABA_stock.csv', delimiter=',',
                       usecols=(0, 1, 2, 3, 4, 5),skiprows=1, dtype='S', converters={0:convert_date})
```

# 股票买卖策略评估

策略: 股价超过10日均线买入, 跌破10日均线卖出   
策略二:10日线上穿60日线买入, 下穿60日线卖出

```python
# 加载数据, 把date这一列设置为索引, 简单起见, 只用收盘价进行分析
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'E:\项目\BABA_stock.csv',index_col='date', 
                usecols=[0,1])
df.head()
```

```python
# 将索引转换为datetime形式
df.index = pd.DatetimeIndex(df.index.str.strip("''"))
df.index
```

```python
# 数据中最近的日期排在前面, 按照日期重新排序
df.sort_index(inplace=True)
print(df.head())
df.describe()
```

```python
# 策略一: 股价超出10日均线买入, 跌破10日均线卖出
# 先计算10均线数据
ma10 = df.rolling(10).mean().dropna()
# 买点
ma10_model = df['close'] - ma10['close']> 0
ma10_model
```

```python
# 第一个值False, 第二个值是True, 在True的时候买入, 需要自定义一个移动窗口处理函数
# 因为卖的时候还需要定义类似的函数, 所以在这里就把两个函数放到一起
# 可以在自定义函数中print一些信息, 例如w值,以方便调试  ----- 这也是调试的一种方式
def get_deal_date(w, is_buy=True):
    if is_buy == True:
        return True if w[0] == False and w[1] == True else False
    else:
        return True if w[0] == True and w[1] == False else False
    
```

```python
# raw= False 没有的话会有警告信息
# 如果删除NA值,会有缺失, 所以这里用0 填充, 转换bool值方便后面取值
se_buy = ma10_model.rolling(2).apply(get_deal_date, raw=False, args=[False]).fillna(0).astype('bool')
se_sale = ma10_model.rolling(2).apply(get_deal_date, raw=False, args=[False]).fillna(0).astype('bool')
# 具体的买卖点
buy_info = df[se_buy.values]
sale_info = df[se_sale.values]
print(buy_info.head())
print(sale_info.head())
```

```python
# 买和卖的索引值不一样, 不过数据都有63条, 所以删除时间索引信息
no_index_buy_info = buy_info.reset_index(drop=True)
no_index_sale_info = sale_info.reset_index(drop= True)
print(no_index_buy_info.head())
print(no_index_sale_info.head())
```

```python
# 每次交易的盈利情况
profit = no_index_sale_info - no_index_buy_info
print(profit)
# 大体数据
profit.describe()
```

```python
profit.sum()
```

```python
# 总利润是36.07, 注意这是每次买和卖一股的利润(买固定的股数),三年的时间交易了63次
#最多投入210.86,平均投入143.3666954, 按最高投入算利润率(36.07/210.86),年化差不多5%, 按平均投入算0.2515,年化将近8%,当然还有手续费没有算
```

```python
# 假如有1w美元，最终盈利多少
all_money = 10000
remain = all_money
for i in range(len(no_index_buy_info)):
    buy_count = remain / no_index_buy_info.iloc[i]
    remain = buy_count * no_index_sale_info.iloc[i]
    print(remain)
    
```
